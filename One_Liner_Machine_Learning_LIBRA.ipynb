Let's install the library. For more info check out [LIBRA](https://github.com/Palashio/libra).

```
!pip install -U libra
```

Let's download the Mobile Phone Price multiclass classification dataset

```
!gdown --id 17wpxB4XilNQ54Yf7aYrpK6vTrvE7f5Gh 
```

Let's pass the dataset in the client, this is the only thing you need to worry about! :)

```
from libra import client 
a_client = client('/content/train.csv') 
```

##**Support Vector Machine**

Let's fit a SVM query to our dataset.

```
a_client.svm_query('model the price_range column', kernel='linear', gamma='scale') 
```

We just need to write the instruction to perform and the function will do all heavy lifting for us. THIS IS AMAZING.

Now to analyze ROC curve, Confusion Matrix and Complete Classification Report, we just call `.analyze` function in libra.

```
a_client.analyze() 
```

Accuracy Metric

```
a_client.accuracy()
```

```
#this is all the information that the query generated
a_client.info()  
```

##**XGBoost**

```
a_client.xgboost_query('please estimate price_range')
```

```
a_client.analyze() 
```

```
a_client.accuracy() 
```
